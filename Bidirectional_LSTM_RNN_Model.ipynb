```python
import tensorflow as tf

from tensorflow.keras.layers import Embedding                     # Helps us to implement word2vec
from tensorflow.keras.preprocessing.sequence import pad_sequences         # Pre & Post padding
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Bidirectional
```

```python
# Assigning the value size to the vocabulary
vocabulary_size=5000
```

One Hot Encoding will take the text data into a vector form and do the feature representation on the data, Where the vocabulary size is 5000 in our case,the words in the data it will arrange the position as per vocabulary size.

# **One-Hot Encoding Representation**

```python
message=X.copy()
```

```python
message
```

# **Let's Clean the Text Data ⏺**

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download("stopwords")
```

```python
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer ##stemming purpose to implement the sentimental analysis
ps = PorterStemmer()
corpus = []
for i in range(0, len(message)):
    review = re.sub('[^a-zA-Z]', ' ', message['title'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

One - Hot Representation

```python
corpus[1]
```

```python
onehot_rep=[one_hot(words,vocabulary_size) for words in corpus]
```

```python
onehot_rep
```

```python
corpus[1]
```

```python
onehot_rep[1]
```

Embedding Representation

```python
sentence_length=20
embedded_doc=pad_sequences(onehot_rep,padding="pre",maxlen=sentence_length)
```

```python
print(embedded_doc)
```

```python
embedded_doc[1]
```

```python
embedded_doc
```

# **Creating the Model and Applying the LSTM & Embedding Layers ❤**

```python
embedding_vector_features=40
model=Sequential()
model.add(Embedding(vocabulary_size,embedding_vector_features,input_length=sentence_length))
model.add(Bidirectional(LSTM(100)))
model.add(Dense(1,activation="sigmoid"))
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
print(model.summary())
```

```python
len(embedded_doc),y.shape
```

```python
X_final=np.array(embedded_doc)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
# Splitting the data into training & Testing
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_final,y_final,test_size=0.30,random_state=45)
```

```python
# Model Training
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=40)
```

```python
# Predicting the model
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred>0.5,1,0)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
